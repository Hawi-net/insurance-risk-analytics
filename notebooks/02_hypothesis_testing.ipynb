```python
import sys
import os

# Ensure the notebook can see the src/ folder
sys.path.append(os.path.abspath(os.path.join('..')))

from src.eda_utils import load_insurance_data
from src.hypothesis_tests import test_numerical_groups

# Load your data safely
df = load_insurance_data("../data/insurance_data.csv")
```

```python
# Testing if TotalClaims are significantly different between Somali and Amhara
prov_test = test_numerical_groups(df, 'Province', 'TotalClaims', 'Somali', 'Amhara')

print("Somali Claim Mean: " + str(prov_test['group1_mean']))
print("Amhara Claim Mean: " + str(prov_test['group2_mean']))
print("P-Value: " + str(prov_test['p_value']))
print("Is the difference statistically significant? " + str(prov_test['significant']))
```

```python
gender_risk_test = test_numerical_groups(df, 'Gender', 'TotalClaims', 'Male', 'Female')

print("Male Claim Mean: " + str(gender_risk_test['group1_mean']))
print("Female Claim Mean: " + str(gender_risk_test['group2_mean']))
print("P-Value: " + str(gender_risk_test['p_value']))
print("Is gender risk difference significant? " + str(gender_risk_test['significant']))
```

### Hypothesis Testing Structural Interpretation
* **Understanding P-Values:** If our calculated `p-value` is **less than 0.05**, we reject the Null Hypothesis. This means the trend we noticed is an actual, statistically proven pattern, not a random fluke. If it is **greater than 0.05**, the difference is likely random variation.
* **Province Risk Analysis:** The test compares the claim variations across geographic profiles to determine if localized premium restructuring is statistically justified.

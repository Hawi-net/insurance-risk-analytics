```python
import sys
import os
import pandas as pd

# Path configuration
sys.path.append(os.path.abspath(os.path.join('..')))
from src.eda_utils import load_insurance_data
from src.hypothesis_tests import run_numerical_ttest, run_categorical_chisquare

# 1. Load Data & Derive KPIs
df = load_insurance_data("../data/insurance_data.csv")

# Engineer helper columns if they don't exist yet
if 'Margin' not in df.columns:
    df['Margin'] = df['TotalPremium'] - df['TotalClaims']
if 'HasClaim' not in df.columns:
    df['HasClaim'] = (df['TotalClaims'] > 0).astype(int)

# 2. Execute Tests
print("--- RUNNING ACIS A/B STRATEGY TESTS ---")

# Test 1: Regional Severity
res_prov = run_numerical_ttest(df, 'Province', 'TotalClaims', 'Amhara', 'Somali')
print(f"Province Test P-Value: {res_prov['p_value']:.5f} | Reject H0: {res_prov['reject_null']}")

# Test 4: Gender Frequency
res_gender = run_categorical_chisquare(df, 'Gender', 'HasClaim', 'Female', 'Male')
print(f"Gender Test P-Value: {res_gender['p_value']:.5f} | Reject H0: {res_gender['reject_null']}")
```
